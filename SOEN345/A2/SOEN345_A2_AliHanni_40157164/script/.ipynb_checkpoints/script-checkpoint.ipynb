```python
data = dict()
file = open('mutations.csv', 'r')
lines = file.readlines()
```

```python
output_data = []
for line in lines:
    package_name = line.split(',')[1].split('.')[4]
    if package_name == 'output':
        output_data.append(line)
```

```python
countKilled = 0
countSurvived = 0
for line in output_data:
    info = line.split(',')
    mutant = info[2].split('.')[6]
    status = info[5]
    if status == 'KILLED':
        countKilled += 1
        data[mutant] = {'SURVIVED' : countSurvived, 'KILLED' : countKilled, 'RATIO' : 0}
    else:
        countSurvived+=1
        data[mutant] = data[mutant] = {'SURVIVED' : countSurvived, 'KILLED' : countKilled, 'RATIO' : 0}
```

```python
for key in data:
    total = data[key]['SURVIVED'] + data[key]['KILLED']
    ratio = data[key]['SURVIVED'] / total
    data[key]['RATIO'] = ratio
```

```python
with open('report.txt', 'w') as report:
    for key, value in data.items():
        report.write('%s: %s\n' % (key, value))
```
