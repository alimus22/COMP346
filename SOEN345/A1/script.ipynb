```python
import json
import xmltodict
```

```python
data = dict()
file = open('data/clover.xml', 'r')
xml_file = file.read()
xml_data = xmltodict.parse(xml_file)
packages = xml_data['coverage']['project']['package']

for package in packages:
    packageName = package['@name']
    files = package['file']
    for file in files:
        name = package['@name'] + '.' + file['@name']
        data[name] = {'JaCoCo': 0, 'Clover': file['metrics']['@coveredstatements']}
```

```python
data
```

```python
file = open('data/jacoco.csv', 'r')
next(file)
lines = file.readlines()
for line in lines:
    info = line.split(',')
    class_name = info[2].split('.')[0]
    name = info[1] + '.' + class_name + '.java'
    data[name]['JaCoCo'] = info[8]
```

```python
data
```

```python
with open('report.txt', 'w') as report:
    for key, value in data.items():
        report.write('%s: %s\n' % (key, value))
```
